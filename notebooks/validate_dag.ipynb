```python
import networkx as nx
import matplotlib.pyplot as plt

# Your json_list goes here
# json_list = [...] 

# 1. Build the Directed Graph
G = nx.DiGraph()
node_labels = {'ROOT': 'ROOT'}
node_times = {'ROOT': 0}

for row in json_list:
    p = row['parent_id']
    c = row['child_id']
    G.add_edge(p, c)
    
    # Store attributes for clean display
    label = row['child_label']
    # If the label is Moderator, we can look at the raw json in your logs 
    # to know what it actually was based on the timestamp/causality, 
    # but for this script we'll just append the short ID to distinguish them.
    node_labels[c] = f"{label} ({c[:8]})"
    node_times[c] = row['updated_at']

print("=========================================")
print(" 🧬 DAG TOPOLOGY VALIDATION")
print("=========================================")

# Test 1: Is it actually a valid Directed Acyclic Graph?
is_dag = nx.is_directed_acyclic_graph(G)
print(f"✅ Is strictly acyclic (Valid DAG): {is_dag}")
print(f"✅ Total Nodes: {G.number_of_nodes()}")
print(f"✅ Total Edges: {G.number_of_edges()}")

# Test 2: Find the exact Forks (nodes that spawned multiple children)
print("\n🔍 Detecting Forks (Out-degree > 1):")
forks = [n for n, d in G.out_degree() if d > 1]
for f in forks:
    children = list(G.successors(f))
    print(f"  ⚠️ Node {node_labels.get(f)} branched into {len(children)} paths:")
    for child in children:
        print(f"      -> {node_labels.get(child)}")

# Test 3: Print the actual topological sequence
print("\n🌳 Topological Spacetime Trace:")
def print_tree(node, depth=0):
    indent = "   " * depth
    print(f"{indent}↳ {node_labels.get(node)}")
    
    # Sort children chronologically to show the exact sequence of events
    children = sorted(list(G.successors(node)), key=lambda c: node_times.get(c, 0))
    for child in children:
        print_tree(child, depth + 1)

print_tree('ROOT')

# # Test 4: Visual Plot
# plt.figure(figsize=(10, 8))
# # Use a spring layout but emphasize the top-down nature
# pos = nx.spring_layout(G, seed=42, k=0.5) 
# nx.draw(G, pos, with_labels=False, node_size=500, node_color="#4ade80", edge_color="#6b7280", width=2, arrows=True)

# # Add custom labels
# for node, (x, y) in pos.items():
#     plt.text(x, y+0.05, node_labels.get(node).split()[0], fontsize=9, ha='center')

# plt.title("Mathematical Edge Connections (Independent of UI Y-Axis)")
# plt.show()

# Test 4: Visual Plot (Chronologically Ordered)
plt.figure(figsize=(12, 10))

# 1. Get an initial layout for the X coordinates to spread out branches
pos = nx.spring_layout(G, seed=42, k=0.8)

# 2. Sort nodes strictly by their timestamp
# ROOT has a time of 0 (from your dictionary setup), so it will naturally be first.
sorted_nodes = sorted(G.nodes(), key=lambda n: node_times.get(n, 0))

# 3. Override the Y-coordinate based on chronological rank
# This ensures time flows evenly from top (oldest) to bottom (newest)
total_nodes = len(sorted_nodes)
for i, node in enumerate(sorted_nodes):
    # Y goes from 1.0 (top) down to 0.0 (bottom)
    y_val = 1.0 - (i / max(1, total_nodes - 1))
    
    # Keep the X coordinate from the spring layout, but force the Y coordinate
    pos[node] = (pos[node][0], y_val)

# Draw the graph
nx.draw(G, pos, with_labels=False, node_size=500, node_color="#4ade80", 
        edge_color="#6b7280", width=2, arrows=True)

# Add custom labels, slightly offset so they don't cover the nodes
for node, (x, y) in pos.items():
    plt.text(x, y + 0.02, node_labels.get(node).split()[0], fontsize=9, ha='center')

plt.title("Spacetime Topology (Y-Axis Ordered by Timestamp Rank)")
plt.show()
```
